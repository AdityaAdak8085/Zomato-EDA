```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/zomato.csv")
df.head()
```

```python
df.drop("menu_item",axis = 1,inplace = True )
```

**Data** **Cleaning**

```python
df.rename(columns = {"phone":"Phone_no","rate":"rating","name":"Hotel_name"},inplace = True)
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df[df["location"]!=df["listed_in(city)"]].size
```

```python
duplicated_rows = df[df.duplicated()]
duplicated_rows.size
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['rating'].fillna(0,inplace = True)
```

```python
df["rating"] = df['rating'].str.replace("/5","")
```

```python
df.head()
```

```python
df.rename(columns ={"rating":"rating(scale = 5)"}).head()
```

```python
df.drop(["Phone_no","dish_liked"],inplace = True,axis = 1)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].str.replace(",","")
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype(float)
```

```python
df["approx_cost(for two people)"].isnull().sum()
```

```python
df["approx_cost(for two people)"]
```

```python
df["approx_cost(for two people)"].mean()
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].fillna(555.4315664479959)
```

```python
df["approx_cost(for two people)"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['location'].mode()
```

```python
df['location'] = df['location'].fillna("BTM")
```

```python
df.head()
```

```python
df['rest_type'].value_counts().idxmax()
```

```python
import matplotlib.pyplot as plt
# plt.figure(figsize=(12,3))

ax=sns.countplot(x='listed_in(type)',data=df)
plt.xticks(rotation=90)
ax.set_xlabel('Types of Restaurants')
ax.set_ylabel('No. of Restaurants')
```

```python
rest_types_rate = pd.DataFrame(df.groupby('listed_in(type)')['rating'].mean())
rest_types_rate.sort_values(by='rating',inplace=True,ascending=False)
ax=sns.barplot(x=rest_types_rate.index, y='rating',data=rest_types_rate)

plt.xticks(rotation=90);
ax.set_title('Restaurant type vs Rating')
ax.set_xlabel('Types of Restaurants')
ax.set_ylabel('Sum of Rating')
```

```python
rest_types_vote=pd.DataFrame(df.groupby('listed_in(type)')['votes'].count())
rest_types_vote.sort_values(by='votes',inplace=True,ascending=False)
ax=sns.barplot(x=rest_types_vote.index,y='votes',data=rest_types_vote)

plt.xticks(rotation=90);
ax.set_title('Restaurant type vs votes')
ax.set_xlabel('Types of Restaurants')
ax.set_ylabel('Total votes')
```

```python
city_amount=pd.Series(df.groupby('listed_in(city)')['approx_cost(for two people)'].mean())
city_amount=city_amount.sort_values(ascending=False)
city_amount.plot(kind='bar',ylabel='Cost for 2 plates',title="City vs Cost of 2 plates")
```
